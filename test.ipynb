# first import pandas library and read csv file 


```python
import pandas as pd
from babel.numbers import format_currency
data = pd.read_csv('test.csv')
data.head()
```

# Count of products without prices and with prices in each Product Type

```python
dff = data[data['price_string'].isnull() == False]
dff.head()
```

```python
arr = dff.product_type.unique()
for each in arr:
    print(dff['product_type'].value_counts())
```

# Count of products without prices and with prices in each Level_1

```python
arr = dff.product_type.unique()
for each in arr:
    print(dff['level_1'].value_counts())
```

# Count of products without prices and with prices in each category

```python
arr = dff.product_type.unique()
for each in arr:
    print(dff['category'].value_counts())
```

```python
dff.head()
```

```python
dff.price_string = dff.price_string.replace({'\$':''}, regex = True)
```
